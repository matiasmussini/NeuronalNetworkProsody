```python
import os
import pandas as pd

# Definir el path de la carpeta que contiene los CSVs
path_csvs = "csvs"

# Crear una lista para almacenar los DataFrames
dataframes = []

# Iterar sobre todos los archivos en la carpeta csvs
for file_name in os.listdir(path_csvs):
    # Procesar solo archivos CSV
    if file_name.lower().endswith('.csv'):
        # Construir el path completo del CSV
        csv_path = os.path.join(path_csvs, file_name)

        try:
            # Leer el archivo CSV en un DataFrame de pandas
            df = pd.read_csv(csv_path)

            # Seleccionar solo las columnas necesarias
            if {'audio_quality', 'vector', 'nombre'}.issubset(df.columns):
                df = df[['audio_quality', 'vector', 'nombre']]
                # A침adir el DataFrame a la lista
                dataframes.append(df)
            else:
                print(f"Advertencia: El archivo {csv_path} no contiene todas las columnas necesarias.")
        except Exception as e:
            print(f"Error al leer el archivo CSV {csv_path}: {e}")

# Concatenar todos los DataFrames en uno solo
if dataframes:
    df_combinado = pd.concat(dataframes, ignore_index=True)
    # Guardar el DataFrame combinado en un nuevo archivo CSV
    df_combinado.to_csv("datos_combinados.csv", index=False)
    print("Todos los archivos CSV han sido combinados en 'datos_combinados.csv'.")
else:
    print("No se encontraron archivos CSV v치lidos para combinar.")
```

Comprobar que todos los vectores tienen la misma longitud

```python
import pandas as pd

# Definir el path del archivo combinado
path_combinado = "datos_combinados.csv"

# Leer el archivo combinado en un DataFrame de pandas
df_combinado = pd.read_csv(path_combinado)

# Definir el vector de referencia
vector_referencia = [0, 1, 0, 0, 0, 1, 1, 0, 0]
longitud_referencia = len(vector_referencia)

# Comprobar si la columna 'vector' y 'nombre' existen en el DataFrame
if 'vector' in df_combinado.columns and 'nombre' in df_combinado.columns:
    # Funci칩n para verificar la longitud del vector
    def longitud_vector_correcta(vector):
        # Convertir el vector a una lista de enteros y verificar su longitud
        try:
            vector_lista = list(map(int, vector.strip('[]').split(',')))
            return len(vector_lista) == longitud_referencia
        except ValueError:
            return False

    # Aplicar la funci칩n a la columna 'vector' y filtrar los vectores incorrectos
    vectores_invalidos = df_combinado[~df_combinado['vector'].apply(longitud_vector_correcta)]

    if not vectores_invalidos.empty:
        # Imprimir el nombre de cada fila con problemas
        print("Filas con problemas en la columna 'vector':")
        for nombre in vectores_invalidos['nombre']:
            print(nombre)
    else:
        print("Todos los vectores tienen la longitud correcta.")
else:
    print("El archivo no contiene las columnas necesarias ('vector' y 'nombre').")
```

```python
import pandas as pd

# Definir el path del archivo combinado
path_combinado = "datos_finales.csv"

# Leer el archivo combinado en un DataFrame de pandas
df = pd.read_csv(path_combinado)
```

Ordenar numeros del 1 al 25

```python
# Ordenar la columna 'nombre' de forma natural para que los nombres est칠n del 1 al 25
# Para lograrlo, podemos usar una ordenaci칩n natural que tiene en cuenta los n칰meros en las cadenas

# Funci칩n para extraer los n칰meros de las cadenas y usarlos para ordenar
import re

def natural_sort_key(s):
    return [int(text) if text.isdigit() else text.lower() for text in re.split('(\d+)', s)]

# Ordenar el dataframe basado en la columna 'nombre'
df_sorted = df.sort_values(by='nombre', key=lambda x: x.map(natural_sort_key))

# Mostrar el resultado ordenado
df_sorted.head(30)  # Mostrar m치s filas para ver c칩mo qued칩 el orden
```

```python
df_sorted.to_csv("datos_finales_ordenados.csv", index=False)
```

detectar audios en los que hay error

```python


# Filtrar las filas donde la columna 'audio_quality' es 'error'
df_error = df_sorted[df_sorted['audio_quality'] == 'Error']

# Obtener los valores de la columna 'nombre' en esas filas
nombre_error = df_error['nombre'].tolist()

# Mostrar los nombres
nombre_error
```

Borrar audios con error

```python
df = df_sorted[~df_sorted['nombre'].isin(nombre_error)]
```

```python
df.to_csv("datos_finales_sin_error.csv", index=False)
```
