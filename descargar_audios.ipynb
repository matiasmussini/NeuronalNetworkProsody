```python
import os
import shutil

# Obtener la lista de carpetas en el directorio actual
carpetas = [f for f in os.listdir('.') if os.path.isdir(f)]

# Iterar sobre las carpetas numeradas del 1 al 25
for i in range(1, 26):
    carpeta_origen = str(i)
    carpeta_destino = None
    
    # Buscar la carpeta de destino que comienza con el mismo número
    for carpeta in carpetas:
        if carpeta.startswith(f"{i}_"):
            carpeta_destino = carpeta
            break
    
    # Si la carpeta de destino fue encontrada
    if carpeta_destino:
        # Obtener la lista de archivos en la carpeta de origen
        archivos_a_mover = [f for f in os.listdir(carpeta_origen) if f.startswith("audio_") and f.endswith(".wav")]
        
        # Mover los archivos correspondientes del 54 al 76
        for archivo in archivos_a_mover:
            num_audio = int(archivo.split('_')[1].split('.')[0])
            if 54 <= num_audio <= 76:
                ruta_origen = os.path.join(carpeta_origen, archivo)
                ruta_destino = os.path.join(carpeta_destino, archivo)
                shutil.move(ruta_origen, ruta_destino)
                print(f"Movido {archivo} de {carpeta_origen} a {carpeta_destino}")
    else:
        print(f"No se encontró carpeta de destino para {carpeta_origen}")
```
