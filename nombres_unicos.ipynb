Nombres Ãºnicos generar

```python
import os

# Definir el path de la carpeta que contiene todas las carpetas de frases
path_principal = "frases"

# Iterar sobre todas las carpetas dentro de la carpeta principal
for folder_name in os.listdir(path_principal):
    # Obtener el nÃºmero de la carpeta al inicio del nombre
    folder_number = folder_name.split('_')[0]

    # Construir el path completo de la carpeta
    folder_path = os.path.join(path_principal, folder_name)

    # Asegurarse de que es una carpeta
    if os.path.isdir(folder_path):
        # Iterar sobre todos los archivos dentro de la carpeta
        for file_name in os.listdir(folder_path):
            # Solo renombrar archivos que empiecen con "audio_"
            if file_name.startswith("audio_"):
                # Construir el nuevo nombre del archivo con el nÃºmero de la carpeta al inicio
                new_file_name = f"{folder_number}{file_name}"

                # Obtener el path completo del archivo actual y del archivo renombrado
                old_file_path = os.path.join(folder_path, file_name)
                new_file_path = os.path.join(folder_path, new_file_name)

                # Renombrar el archivo
                os.rename(old_file_path, new_file_path)

                print(f'Renombrado: {old_file_path} -> {new_file_path}')

print("Renombrado completado.")
```

Comprobar que estÃ¡ bien hecho

```python
import os

# Definir el path de la carpeta que contiene todas las carpetas de frases
path_principal = "frases"

# Crear un conjunto para almacenar los nombres de archivos que deberÃ­an existir
archivos_esperados = set()

# Generar los nombres esperados para todos los archivos
for i in range(1, 26):
    for j in range(2, 77):
        archivos_esperados.add(f"{i}audio_{j}.wav")

# Crear un conjunto para almacenar los nombres de archivos que realmente existen
archivos_encontrados = set()

# Iterar sobre todas las carpetas dentro de la carpeta principal
for folder_name in os.listdir(path_principal):
    # Obtener el nÃºmero de la carpeta al inicio del nombre
    folder_number = folder_name.split('_')[0]

    # Construir el path completo de la carpeta
    folder_path = os.path.join(path_principal, folder_name)

    # Asegurarse de que es una carpeta
    if os.path.isdir(folder_path):
        # Iterar sobre todos los archivos dentro de la carpeta
        for file_name in os.listdir(folder_path):
            if file_name.endswith(".wav"):
                # AÃ±adir el archivo al conjunto de archivos encontrados
                archivos_encontrados.add(file_name)

# Comparar los archivos encontrados con los archivos esperados
archivos_faltantes = archivos_esperados - archivos_encontrados
archivos_extra = archivos_encontrados - archivos_esperados

# Imprimir resultados
if not archivos_faltantes and not archivos_extra:
    print("Todos los archivos estÃ¡n presentes y correctos.")
else:
    if archivos_faltantes:
        print("Faltan los siguientes archivos:")
        for archivo in sorted(archivos_faltantes):
            print(archivo)
    if archivos_extra:
        print("\nExisten archivos adicionales no esperados:")
        for archivo in sorted(archivos_extra):
            print(archivo)
```

Pasar todos los audios a una misma carpeta

```python
import os
import shutil

# Definir el path de la carpeta que contiene todas las carpetas de frases
path_principal = "frases"

# Definir el path de la carpeta de destino donde se moverÃ¡n todos los audios
carpeta_destino = os.path.join(path_principal, "todos_los_audios")

# Crear la carpeta de destino si no existe
os.makedirs(carpeta_destino, exist_ok=True)

# Iterar sobre todas las carpetas dentro de la carpeta principal
for folder_name in os.listdir(path_principal):
    # Construir el path completo de la carpeta
    folder_path = os.path.join(path_principal, folder_name)

    # Asegurarse de que es una carpeta
    if os.path.isdir(folder_path) and folder_name != "todos_los_audios":
        # Iterar sobre todos los archivos dentro de la carpeta
        for file_name in os.listdir(folder_path):
            if file_name.endswith(".wav"):
                # Construir el path completo del archivo y su destino
                old_file_path = os.path.join(folder_path, file_name)
                new_file_path = os.path.join(carpeta_destino, file_name)

                # Mover el archivo a la carpeta de destino
                shutil.move(old_file_path, new_file_path)

                print(f'Movido: {old_file_path} -> {new_file_path}')

print("Todos los archivos se han movido a la carpeta 'todos_los_audios'.")
```
